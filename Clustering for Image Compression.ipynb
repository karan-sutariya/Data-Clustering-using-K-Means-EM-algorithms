# <font color='mediumblue' face='times new roman' size='6.5'>K-Means Clustering for Image Compression or Image Segmentation</font>

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from skimage import io as img
pic = img.imread('https://upload.wikimedia.org/wikipedia/commons/a/a1/Cnidoscolus_stimulosus_%28flower%29_low_res.jpg?20110217152652')
#plotting the image
img.imshow(pic.astype(np.uint8))
img.show()
```

# <font color='crimson' face='times new roman' size='5'> K-Means Algorithm </font>

```python
#Assigning each datapoint to the nearest centroid
def cluster_assignment(data, centroids):
    distances = np.linalg.norm(data[:, np.newaxis, :] - centroids, axis=2)
    clusters = np.argmin(distances, axis=1)
    return clusters

#Updating the centers of each clusters
def centroids_update(data, clusters, K):
    new_centroids = np.zeros_like(data[:K])
    for u in range(K):
        cluster_points = data[clusters==u]
        if(len(cluster_points) > 0):
            new_centroids[u] = np.mean(cluster_points, axis=0)
    return new_centroids #Assigning each datapoint to the nearest centroid

#Assignment of clusters to each point
def cluster_assignment(data, centroids):
    distances = np.linalg.norm(data[:, np.newaxis, :] - centroids, axis=2)
    j = np.argmin(distances, axis=1)
    return j

def K_MEANS(data, K):
    epochs = 1000
    centroids = data[np.random.choice(data.shape[0], K, replace=False)]

    for itr in range(epochs):
        clusters = cluster_assignment(data, centroids)
        new_centroids = centroids_update(data, clusters, K)

        if np.all(centroids == new_centroids):
            break

        centroids = new_centroids

    return clusters, centroids
```

# <font color='crimson' face='times new roman' size='5'>Image Flattening</font>

```python
# Reshape the image into a flattened form
num_rows, num_columns, num_channels = pic.shape
flattened_img = pic.reshape(num_rows * num_columns, -1)
```

# <font color='blue' face='times new roman' size='5'>Applying K-Means for different values of K</font>

# <font color='darkRED' face='times new roman' size='5'>For K = 5</font>

```python
# Apply K-Means clustering to find 5 clusters
classes_1, centroids_1 = K_MEANS(flattened_img, 5)

# Compress the image using the cluster centers and labels
flattened_compressed_image_1 = centroids_1[classes_1]

# Reshape the compressed image back to its original shape
compressedImage1 = flattened_compressed_image_1.reshape(num_rows, num_columns, -1)

img.imshow(compressedImage1.astype(np.uint8))
img.show()
```

# <font color='darkRED' face='times new roman' size='5'>For K = 10</font>

```python
# Apply K-Means clustering to find 10 clusters
classes_2, centroids_2 = K_MEANS(flattened_img, 10)

# Compress the image using the cluster centers and labels
flattened_compressed_image_2 = centroids_2[classes_2]

# Reshape the compressed image back to its original shape
compressedImage2 = flattened_compressed_image_2.reshape(num_rows, num_columns, -1)

img.imshow(compressedImage2.astype(np.uint8))
img.show()
```
